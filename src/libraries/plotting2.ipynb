```python
import matplotlib.pyplot as plt
import pandas as pd
print('hello world')
```

## Load data

```python
df = pd.read_csv('Data/Agrofood_co2_emission.csv',sep=',')
print(df.head())
```

## Line plot

```python
zdf = df[df['Area'] == 'Brazil']
cdf = df[df['Area'] == 'Canada']
# print(zdf)

fig, ax = plt.subplots()
ax.plot(zdf['Year'],zdf['Fires in humid tropical forests'],label='Brazil')
ax.plot(cdf['Year'],cdf['Fires in humid tropical forests'], label='Canada')
ax.legend(loc='upper right')
ax.set_xlabel('Year')
ax.set_ylabel('Fires in humid tropical forests')
plt.show()
```

## histogram

```python
fig, ax = plt.subplots()
ax.hist(zdf['Savanna fires'],bins=20,edgecolor='black')
ax.set_xlabel('Savanna fires')
ax.set_ylabel('Freq.')
ax.set_title('A',loc='left')
ax.spines[['right', 'top']].set_visible(False)
print(ax.get_xlim())
ax.hlines(y=3,xmin=ax.get_xlim()[0],xmax=ax.get_xlim()[1],color='red',linestyle='--')

plt.show()
```

## Scatter

```python
import numpy as np

fig, ax = plt.subplots()
ax.scatter(zdf['Pesticides Manufacturing'],zdf['total_emission'],alpha=.5,s=50,c='#e65100')

plt.plot(np.unique(zdf['Pesticides Manufacturing']), np.poly1d(np.polyfit(zdf['Pesticides Manufacturing'], zdf['total_emission'], 1))(np.unique(zdf['Pesticides Manufacturing'])))

plt.show()
```

## Multipanel plots

```python
import seaborn as sns
plt.rcParams.update({'font.size': 22})

fig, axes = plt.subplots(2,2,width_ratios=[1, 1],height_ratios=[1,1])
fig.set_size_inches(10, 10)
print(axes.shape)
axes[0,0].scatter(zdf['Pesticides Manufacturing'],zdf['total_emission'],alpha=.5,s=50,c='#e65100')
axes[0,0].set_title('A',loc='left')
axes[0,0].set_xlabel('Pesticides Manufacturing')
axes[0,0].set_ylabel('Total emission')

axes[0,1].boxplot(zdf['Pesticides Manufacturing'])
axes[0,1].set_title('B\n',loc='left')
axes[0,1].set_ylabel('Pesticides Manufacturing')

axes[1,0].violinplot(zdf['total_emission'])
sns.kdeplot(np.array(zdf['total_emission']), bw=0.5,ax=axes[1,1])

for i in range(axes.shape[0]):
    for j in range(axes.shape[1]):
           axes[i,j].spines[['right', 'top']].set_visible(False)


plt.tight_layout()
plt.savefig('testing_figure.png',dpi=300)

plt.show()
```
